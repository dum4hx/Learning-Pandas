## 5.1 Create new column clasifying people by their generation

```python
def clasify_by_generation(row) -> str:
    """
    # Parameters
        A row from a `pd.DataFrame` class.
    # Return
        A `str` either "`Baby Boomers`", "`X`", "`Milennials`" or "`Z`".
        """
    born_year = row["born_year"]
    
    if  1946 >= born_year <= 1964:
        return "Baby Boomers"
    elif 1964 >= born_year <= 1980:
        return "X"
    elif 1981 >= born_year <= 1996:
        return "Milennials"
    elif 1997 >= born_year <= 2012:
        return "Z"
    else:
        return "Other"
```

```python
df["generation"] = df.apply(clasify_by_generation, axis=1)
```

#### 6.1.1 Prepare nocs dataset

```python
nocs_link = "https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv"
nocs = pd.read_csv(nocs_link)
```

```python
nocs.head()
```

#### 6.1.2 Drop unnecessary columns

```python
nocs = nocs.loc[:, ["name", "alpha-3"]]
```

```python
df[df["NOC"] == "United States"]
```

```python
nocs.loc[nocs["alpha-3"].str.upper() == "USA", "name"] = "United States"
nocs.loc[nocs["alpha-3"].str.upper() == "USA", "name"] = "United States"
```

```python
nocs[nocs["alpha-3"] == "USA"].head()
```

```python
merged_df = pd.merge(df, nocs, left_on="born_country", right_on="alpha-3", how="inner")
```

```python
merged_df[merged_df["name_y"].isna() == True]
```

```python
# Drop Born country
merged_df = merged_df.drop(columns=["born_country"])

# Rename columns
merged_df = merged_df.rename(columns={"name_y": "born_country_full"})
merged_df = merged_df.rename(columns={"name_x": "name"})
merged_df = merged_df.rename(columns={"alpha-3": "born_country"})
```

```python
merged_df.info()
```

```python
merged_df[merged_df["born_country"].str.contains("COL")].head()
```

```python
merged_df[merged_df["born_country_full"] != merged_df["NOC"]][["name", "born_country", "born_country_full", "NOC"]]
```

```python
merged_df.loc[merged_df["NOC"] == "GBR", "born_country_full"] = "Great Britain" 
merged_df.loc[merged_df["born_country_full"] == "Korea, Republic of", "born_country_full"] = "Republic of Korea" 
merged_df.loc[merged_df["born_country_full"] == "Congo, Democratic Republic of the", "born_country_full"] = "Democratic Republic of the Congo" 
```
